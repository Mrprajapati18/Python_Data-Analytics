# Distributions of Features and species

```python
df["SepalLengthCm"].hist()
plt.title('trend the iris dataset')
plt.ylabel('number of ')
plt.xlabel('SepalWidthCm')

plt.show()
```

```python
df["SepalWidthCm"].hist()
plt.title('trend the iris dataset')
plt.ylabel('number of SepalLengthCm')
plt.xlabel('SepalWidthCm')

plt.show()
```

```python
df["Species"].hist()
plt.title('trend the iris dataset Species')
plt.ylabel('number of SepalLengthCm')
plt.xlabel('Species')

plt.show()
```

```python
df["SepalLengthCm"].hist()
plt.title('trend the iris dataset Species')
plt.ylabel('number of Species ')
plt.xlabel('SepalWidthCm')

plt.show()
```

```python
df["PetalLengthCm"].hist()
plt.title('trend the iris dataset ')
plt.ylabel('number of PetalLengthCm')
plt.xlabel('Species')

plt.show()
```

# Explonatory Data Analysis(EDA) pairplot

```python
sns.pairplot(df)

plt.ylabel('number of PetalLengthCm')
plt.xlabel('Species')

plt.show()
```

```python
sns.pairplot(df,hue="Species")
plt.title('trend the iris dataset ')
plt.ylabel('number of Species ')
plt.xlabel('SepalWidthCm')
plt.show()
```
