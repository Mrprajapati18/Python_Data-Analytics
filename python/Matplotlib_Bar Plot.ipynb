# Bar plot

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
plt.bar(x,y)

plt.show()
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
plt.bar(x,y,color="r")

plt.show()
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
plt.bar(x,y,width=0.4)

plt.show()
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
plt.bar(x,y)
d=["y","b","r","m"]
plt.bar(x,y,color=d)
plt.show()
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.bar(x,y)
d=["y","b","r","m"]
plt.bar(x,y,color=d)
plt.show()
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
plt.bar(x,y)
d=["y","b","r","m","g"]
plt.bar(x,y,color=d,edgecolor="r")
plt.show()
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
plt.bar(x,y)
d=["y","b","r","m","g"]
plt.bar(x,y,color=d,edgecolor="b",linewidth=2)
plt.show()
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
plt.bar(x,y)
d=["y","b","r","m","g"]
plt.bar(x,y,color=d,edgecolor="m",linewidth=2,linestyle=":")
plt.show()
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
plt.bar(x,y)
d=["y","b","r","m","g"]
plt.bar(x,y,color=d,edgecolor="m",linewidth=2,linestyle=":",alpha=0.2)#alpha values is 0 to 1
plt.show()
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
plt.bar(x,y)
d=["y","b","r","m","g"]
plt.bar(x,y,color=d,edgecolor="m",label="thatsmycollege.com")
plt.legend()
plt.show()
```

```python
x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
z=[40,50,60,30,55]

plt.xlabel("language")
plt.ylabel("number")
width=0.4
plt.bar(x,y)
plt.bar(x,z,color="g",label="thatsmycollege.com")
plt.bar(x,z,color="y",label="thatsmycollege.com1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","c","c++","java","javascript"]
y=[75,68,98,50,85]
z=[40,50,60,30,55]

width=0.5
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("thatsmycollege",fontsize=15)

plt.bar(p,y,width,color="g",label="thatsmycollege.com")
plt.bar(p1,z,width,color="y",label="thatsmycollege.com1")


plt.xticks(p+width/2,x,rotation=5)# x axis language in center
plt.legend()
plt.show()
```
