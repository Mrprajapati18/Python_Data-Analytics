# Scatter plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
day=[1,2,3,4,5,6,7]
number=[2,4,3,1,5,3,6]
plt.scatter(day,number)
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.show()
```

# change scatter color

```python
day=[1,2,3,4,5,6,7]
number=[2,4,3,1,5,3,6]
plt.scatter(day,number,color="g")
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.show()
```

# Different color

```python
day=[1,2,3,4,5,6,7]
number=[2,4,3,1,5,3,6]
colors=["g","y","r","m","b","g","m"]
plt.scatter(day,number,c=colors)
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.show()
```

# Dot sizes change

```python
day=[1,2,3,4,5,6,7]
number=[2,4,3,1,5,3,6]
colors=["g","y","r","m","b","g","m"]
sizes=[50,54,45,53,51,43,100]
plt.scatter(day,number,c=colors,s=sizes,alpha=0.5)# alpha values is 0 to 1
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.show()
```

# color bar

```python
day=[1,2,3,4,5,6,7]
number=[2,4,3,1,5,3,6]
colors=[10,40,30,20,50,60,40]
sizes=[50,54,45,53,51,43,100]
plt.scatter(day,number,c=colors,cmap="viridis",s=sizes)
plt.colorbar()
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.show()
```

# color bar label

```python
day=[1,2,3,4,5,6,7]
number=[2,4,3,1,5,3,6]
colors=[10,40,30,20,50,60,40]
sizes=[50,54,45,53,51,43,100]
plt.scatter(day,number,c=colors,cmap="BrBG",s=sizes)
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.show()
```

# multiple color bar

```python
day=[1,2,3,4,5,6,7]
number=[2,4,3,1,5,3,6]
number1=[5,3,6,4,2,1,2]
colors=[10,40,30,20,50,60,40]
sizes=[50,54,45,53,51,43,100]
plt.scatter(day,number,c=colors,cmap="BrBG",s=sizes)
plt.scatter(day,number1,c=colors,cmap="viridis",s=sizes)
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("scatter plot",fontsize=15)
plt.xlabel("day",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.show()
```
