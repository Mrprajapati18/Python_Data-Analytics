```python
plt.hist(number,"auto",(0,100),edgecolor="r")
plt.title("dataset")
plt.xlabel("python")
plt.ylabel("No")
plt.show()
```

```python
plt.hist(number,color="g",edgecolor="r",cumulative =-1)
plt.title("dataset")
plt.xlabel("python")
plt.ylabel("No")
plt.show()
```

```python
plt.hist(number,color="b",edgecolor="r",bottom=10)
plt.title("dataset")
plt.xlabel("python")
plt.ylabel("No")
plt.show()
```

```python
plt.hist(number,color="r",edgecolor="g",align="left")
plt.title(" dataset")
plt.xlabel("python")
plt.ylabel("No")
plt.show()
```

```python
plt.hist(number,color="y",edgecolor="g",histtype="step")
plt.title(" dataset")
plt.xlabel("python")
plt.ylabel("No")
plt.show()
```

```python
plt.hist(number,color="y",edgecolor="g",histtype="bar")
plt.title(" dataset")
plt.xlabel("python")
plt.ylabel("No")
plt.show()
```

```python
plt.hist(number,"auto",(0,100),color="r",edgecolor="g",rwidth=0.8)
plt.title(" dataset")
plt.xlabel("python")
plt.ylabel("No")
plt.show()
```

```python
plt.hist(number,color="g",edgecolor="r",label="python",alpha=0.4)
plt.title(" dataset")
plt.xlabel("python")
plt.ylabel("No")
plt.legend()
plt.show()
```
