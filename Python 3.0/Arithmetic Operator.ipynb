Arithmetic Operator

```python
import pandas as pd
```

```python
var = pd.DataFrame({"A":[1,2,3,4,5,6],"B":[7,8,9,10,11,12]})
var
```

```python
var["C"] = var["A"] + var["B"]
var
```

```python
var["C"] = var["A"] - var["B"]
var
```

```python
var["C"] = var["A"] * var["B"]
var
```

```python
var["C"] = var["A"]/ var["B"]
var
```

```python
var["C"] = var["A"] % var["B"]
var
```

Logical Operator

```python
Var_1 = pd.DataFrame({"A": [10,20,30,40,50],"B": [15,16,17,18,19]})
Var_1
```

```python
Var_1["Python"] = Var_1["A"] <= 20
Var_1
```

```python
Var_1["Python_3.0"] = Var_1["B"] >= 16
Var_1
```
