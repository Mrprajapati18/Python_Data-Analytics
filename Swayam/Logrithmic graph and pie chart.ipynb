# import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,20,100)
y=5*x**3
plt.loglog(x,y)
plt.show()
```

# pie chart

```python
x=[10,20,30,40,50]
y=["c","c++","python","java","javascript"]
plt.pie(x)
plt.show()
```

```python
x=[10,20,30,40,50]
y=["c","c++","python","java","javascript"]
ex=[0.3,0.0,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
x=[10,20,30,40,50]
y=["c","c++","python","java","javascript"]
ex=[0.3,0.0,0.0,0.0,0.0]
c=["y","g","r","b","m"]
plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
```

```python
x=[10,20,30,40,50]
y=["c","c++","python","java","javascript"]
ex=[0.3,0.0,0.0,0.0,0.0]
c=["y","g","r","b","m"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%")
plt.show()
```
