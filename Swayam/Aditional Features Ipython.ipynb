```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-2*np.pi,2*np.pi,100)
plt.plot(x,x*np.sin(x))

plt.show()
```

```python
np.linspace(-2*np.pi,2*np.pi,100)
plt.plot(x,x*np.sin(x))
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.title("$x and x sin(x)$")
plt.show()
```
