```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,50,10)
plt.plot(x,np.sin(x))
plt.plot()
```

```python
x=np.linspace(0,50,10)
plt.plot(x,np.sin(x))
y=np.linspace(0,50,500)
plt.plot(y,np.sin(y))
plt.plot()
```

```python
x=np.linspace(0,50,10)
plt.plot(x,np.sin(x))
y=np.linspace(0,50,500)
plt.plot(y,np.sin(y))
plt.legend(['sin(x)','sin(y)'])
plt.plot()
```

```python
x=np.linspace(0,50,10)
plt.plot(x,np.sin(x))
y=np.linspace(0,50,500)
plt.plot(y,np.sin(y))
plt.title("sin graph",fontsize="15")
plt.plot()
```
