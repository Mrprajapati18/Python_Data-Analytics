```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,50,500)
plt.plot(x,np.cos(x))
plt.show()
```

```python
x=np.linspace(0,50,500)
plt.plot(x,np.cos(x))
y=np.linspace(0,50,500)
plt.plot(y,np.sin(y))
plt.legend(['cos(x)','sin(y)'])
plt.show()
```

```python
x=np.linspace(0,50,500)
plt.plot(x,np.cos(x))
y=np.linspace(0,50,100)
plt.plot(y,y**2)
plt.xlabel("$x$")
plt.ylabel("$Y$")
plt.show()
```
