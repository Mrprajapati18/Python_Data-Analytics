# plot graph x*x+4*x-5

```python
x=np.linspace(2,20,5)
plt.plot(x,-x*x+4*x-5)
plt.title(r"parabolic function $ x^x+4^x-5 $")
plt.show()
```

```python
x=np.linspace(2,20,5)
plt.plot(x,-x*x+4*x-5)
plt.title(r"parabolic function $ x^x+4^x-5 $")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.show()
```
